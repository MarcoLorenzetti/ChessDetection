```python
import ChessDetectionClass as cdc
import numpy as np
import cv2
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import os
import pandas as pd
```

**Set Detector**

```python
inputs = {}

# Processing Image
inputs["blur_k"] = 5
inputs["contrast"] = 1.5
inputs["canny_thr_min"] = 65.13
inputs["canny_thr_max"] = 146.65
inputs["apertureSize"] = 3

# Warping
inputs["width"] = 500
inputs["height"] = 500

# Hough transform
inputs["rho"] = 1
inputs["theta_divider"] = 180
inputs["HT_threshold"] = 91.15
inputs["slope_thr"] = 0.06

# Horizontal/Vertical Clustering
inputs["eps"] = 15
inputs["min_samples"] = 2

# Mask Processing
inputs["slope_0"] = 1
inputs["area_thr"] = 2000
inputs["sides_ratio"] = 0.8

# Optimization
inputs["method"] = 'Powell'

DET = cdc.ChessDetector(inputs)
DET
```

```python
DET.ReturnParameters()
```

## Detections

```python
for i in range(1, 7):
    print(f'-----------Chessboard {i}------------')
    
    img_path = f'IMAGES/chessboards/{i}.jpg'
    board, positions, FEN = DET.ScanBoard(img_path,
                                          plot_detection=False, plot_squares=True,
                                          display_board=True, return_fen=True)
    plt.show()
    
    print(f'\nFEN: {FEN}')
```
